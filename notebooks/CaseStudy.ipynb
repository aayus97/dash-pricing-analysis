```python
# Define a function to clean the data

def clean_data(df):
    # Remove empty rows and columns
    df.dropna(how='all', inplace=True)
    df.dropna(axis=1, how='all', inplace=True)
    
    # Standardize column names
    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
    
    # Ensure price columns are numeric
    price_columns = [col for col in df.columns if col not in ['brand', 'name', 'category', 'sku','url']]
    df[price_columns] = df[price_columns].apply(pd.to_numeric, errors='coerce')
    # df[price_columns].fillna(0, inplace=True)  # Optional: Replace NaNs with 0
    
    return df, price_columns
```

```python
# Clean both weeks' data

df_week1, price_columns_week1 = clean_data(df_w_1)
df_week2, price_columns_week2 = clean_data(df_w)
```

```python
# Create DataFrames to store all products with price increases and decreases across all regions
all_price_increases = pd.DataFrame()
all_price_decreases = pd.DataFrame()

# Loop through each region and identify products with price increases or decreases
for region in price_columns_week1:
    diff_col = f"{region}_price_difference"
    
    # Filter products with price increases for the current region
    price_increase = df_merged[df_merged[diff_col] > 0][['sku', diff_col]]
    price_increase['region'] = region
    all_price_increases = pd.concat([all_price_increases, price_increase], ignore_index=True)
    
    # Filter products with price decreases for the current region
    price_decrease = df_merged[df_merged[diff_col] < 0][['sku', diff_col]]
    price_decrease['region'] = region
    all_price_decreases = pd.concat([all_price_decreases, price_decrease], ignore_index=True)

# Display the results
print("Products with price increases across regions:")
print(all_price_increases.head(10))

print("\nProducts with price decreases across regions:")
print(all_price_decreases.head(10))
```

```python
# Calculate Percentage Change in Price

for region in price_columns_week1:
    week1_col = f"{region}_week1"
    week2_col = f"{region}_week2"
    pct_change_col = f"{region}_percentage_change"
    
    # Calculate percentage change
    df_merged[pct_change_col] = ((df_merged[week2_col] - df_merged[week1_col]) / df_merged[week1_col]) * 100
```

```python
# Dictionary to store the top 10 percentage changes for all regions
top_percentage_changes = {}

for region in price_columns_week1:
    pct_change_col = f"{region}_percentage_change"
    
    # Skip regions that have no calculated percentage change column
    if pct_change_col not in df_merged.columns:
        continue
    
    # Sort the dataset by the percentage change column in descending order
    sorted_changes = df_merged[['brand_x', 'name_x', 'sku', pct_change_col]].sort_values(
        by=pct_change_col, ascending=False
    )
    
    # Get the top 10 changes
    top_10 = sorted_changes.head(10)
    top_percentage_changes[region] = top_10

# Print the top 10 percentage changes for all regions
for region, top_10 in top_percentage_changes.items():
    print(f"\nTop 10 Percentage Changes in {region}:")
    print(top_10)
```

```python
# Detect Outliers 
for region in price_columns_week1:
    diff_col = f"{region}_price_difference"
    Q1 = df_merged[diff_col].quantile(0.25)  # First quartile
    Q3 = df_merged[diff_col].quantile(0.75)  # Third quartile
    IQR = Q3 - Q1  # Interquartile range

    # Identify outliers (below Q1 - 1.5*IQR or above Q3 + 1.5*IQR)
    outliers = df_merged[(df_merged[diff_col] < (Q1 - 1.5 * IQR)) | 
                         (df_merged[diff_col] > (Q3 + 1.5 * IQR))]

    # Print the SKU and price difference for the detected outliers
    print(f"\nOutliers in {region}:")
    print(outliers[['sku', diff_col]])
```

```python
import plotly.express as px

# Ensure 'brand' and 'name' are present in df_merged
if 'brand_x' not in df_merged.columns or 'name_x' not in df_merged.columns:
    raise ValueError("The columns 'brand' and 'name' are missing in the dataset.")

# Consolidate all outliers
outliers_all = []
for region in price_columns_week1:
    diff_col = f"{region}_price_difference"
    Q1 = df_merged[diff_col].quantile(0.25)
    Q3 = df_merged[diff_col].quantile(0.75)
    IQR = Q3 - Q1
    region_outliers = df_merged[
        (df_merged[diff_col] < (Q1 - 1.5 * IQR)) | (df_merged[diff_col] > (Q3 + 1.5 * IQR))
    ]
    # Ensure 'brand' and 'name' are included
    region_outliers = region_outliers[['sku', 'brand_x', 'name_x', diff_col]].copy()
    region_outliers['region'] = region
    region_outliers.rename(columns={diff_col: 'Price Difference'}, inplace=True)
    outliers_all.append(region_outliers)

# Combine into a single DataFrame
outliers_combined = pd.concat(outliers_all, ignore_index=True)

# Plot scatter plot
fig = px.scatter(
    outliers_combined,
    x="region",
    y="Price Difference",
    color="brand_x",
    title="Outliers in Price Differences Across Regions",
    labels={"Price Difference": "Price Difference", "brand_x": "Brand"},
    hover_data=["sku", "brand_x", "name_x"]
)
fig.update_traces(marker=dict(size=8))
fig.show()

```

```python
# # Top Movers Analysis
# # top products with the highest and lowest percentage price changes for each region

top_movers = {}

for region in price_columns_week1:
    pct_change_col = f"{region}_percentage_change"
    
    # Check if the column exists in the DataFrame
    if pct_change_col not in df_merged.columns:
        continue  # Skip regions without percentage change data

    # Find top 5 products with highest and lowest percentage change
    top_increase = df_merged.nlargest(5, pct_change_col)[['sku', pct_change_col]]
    top_decrease = df_merged.nsmallest(5, pct_change_col)[['sku', pct_change_col]]
    
    # Optionally, include brand and name if available
    if 'brand' in df_merged.columns and 'name' in df_merged.columns:
        top_increase = df_merged.nlargest(5, pct_change_col)[['sku', 'brand', 'name', pct_change_col]]
        top_decrease = df_merged.nsmallest(5, pct_change_col)[['sku', 'brand', 'name', pct_change_col]]

    top_movers[region] = {'Top Increase': top_increase, 'Top Decrease': top_decrease}

# Example: Print top movers for the first region
first_region = list(top_movers.keys())[0]
print(f"Top Increases in {first_region}:")
print(top_movers[first_region]['Top Increase'])

print(f"\nTop Decreases in {first_region}:")
print(top_movers[first_region]['Top Decrease'])
```

```python
import plotly.express as px

# Consolidate all outliers in USD
outliers_all_usd = []

# Iterate through the columns for price differences in USD
price_diff_usd_columns = [col for col in df_merged.columns if '_price_diff_USD' in col]

for price_diff_col in price_diff_usd_columns:
    # Extract the region name from the column name
    region = price_diff_col.replace('_price_diff_USD', '').replace('_', ' ').title()

    # Calculate IQR for the column
    Q1 = df_merged[price_diff_col].quantile(0.25)
    Q3 = df_merged[price_diff_col].quantile(0.75)
    IQR = Q3 - Q1

    # Identify outliers for the region
    region_outliers = df_merged[
        (df_merged[price_diff_col] < (Q1 - 1.5 * IQR)) | 
        (df_merged[price_diff_col] > (Q3 + 1.5 * IQR))
    ]

    # Check if there are any outliers
    if not region_outliers.empty:
        region_outliers = region_outliers[['sku', 'brand_week1', 'name_week1', price_diff_col]].copy()
        region_outliers['Region'] = region
        region_outliers.rename(columns={price_diff_col: 'Price Difference (USD)'}, inplace=True)
        outliers_all_usd.append(region_outliers)

# Combine all outliers into a single DataFrame
if outliers_all_usd:
    outliers_combined_usd = pd.concat(outliers_all_usd, ignore_index=True)

    # Plot the outliers using Plotly
    fig = px.scatter(
        outliers_combined_usd,
        x="Region",
        y="Price Difference (USD)",
        color="brand_week1",
        title="Outliers in Price Differences Across Regions (USD)",
        labels={"Price Difference (USD)": "Price Difference (USD)", "Region": "Region"},
        hover_data=["sku", "brand_week1", "name_week1"]
    )
    fig.update_traces(marker=dict(size=8))
    fig.show()
else:
    print("No outliers detected across regions.")
```

1. Structuring and Organizing Data Effectively
Objective: Ensure your data is clean, consistently formatted, and logically organized for analysis.
Actions to Take:
Standardize Columns: Ensure all columns have consistent naming conventions and clear labels (e.g., "region", "brand", "price_change").
Convert Data Types: Ensure all numerical data (like prices) are in the correct format for calculations. Use a consistent currency (like USD) if data is in multiple currencies, applying conversion rates early in the process.
Handle Missing Data: Decide on a strategy for handling missing values, either by filling them in with appropriate values (e.g., zero or mean) or by filtering them out if they don’t impact your insights.
2. Generating Insights from Raw Data
Objective: Use organized data to extract meaningful insights.
Actions to Take:
Calculate Key Metrics: For example, compute percentage changes in price, average price differences by region, and total counts of price increases or decreases.
Visualize Trends: Use charts and graphs to uncover trends (e.g., heatmaps for average price changes by brand and region, bar charts for total increases or decreases by region).
Summarize Key Findings: Summarize patterns, such as regions with consistent increases or decreases in prices, brands with significant variability, or categories with unusual price trends.
3. Identifying and Handling Anomalies in Data
Objective: Detect and understand outliers or inconsistencies in your data.
Actions to Take:
Outlier Detection: Use interquartile range (IQR) or other statistical methods to identify outliers in price changes. Document how you are detecting these anomalies (e.g., anything beyond 1.5 * IQR from the median).
Interpret Anomalies: Look for patterns in outliers by region or brand to understand potential causes (e.g., market differences, promotional pricing).
Exclude or Highlight: Decide whether to exclude anomalies from analysis or to highlight them in your final dashboard as unusual data points, explaining the rationale behind your decision.
If these actions do not yield the desired insights or clarity in your data, consider:
Reviewing Data Completeness: Ensure all relevant data (such as week-over-week data) is correctly imported and merged without loss of information.
Adjusting Data Transformation Steps: Check the transformations (like currency conversion) for accuracy and appropriateness.
Re-evaluating Visualization Choices: Choose visualizations that directly address your insights, such as switching from line plots to box plots for showing distribution.
Each of these steps should help refine the data analysis to achieve the intended results in identifying key insights, handling anomalies, and structuring data effectively for decision-making.
